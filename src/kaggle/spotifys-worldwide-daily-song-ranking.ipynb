```python
# Find the top 10 ranked songs by position. 
# Output the track name along with the corresponding position and sort records by the position in descending order and track name alphabetically, as there are many tracks that are tied for the same position.

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
spotify_df = pd.read_csv('/kaggle/input/spotifys-worldwide-daily-song-ranking/data.csv')
```

```python
spotify_df.head()
```

```python
spotify_df = spotify_df.sort_values(by = ['Streams', 'Track Name'], ascending = [False, True])
```

```python
spotify_df
```

```python
spotify_df = spotify_df.drop_duplicates(subset = ['Track Name'], keep = 'first')
```

```python
top_10 = spotify_df.head(10)
```

```python
top_10
```

```python
top_10.insert(0, 'Rank', value = range(1, 11))
```

```python
top_10_result = top_10.loc[:, ['Rank', 'Track Name']]
```

```python
top_10_result
```

```python
# Alternative: get top 10 songs per region, sort and remove duplicates

# get the top10 songs
spotify_df = pd.read_csv('/kaggle/input/spotifys-worldwide-daily-song-ranking/data.csv')

top10_songs = spotify_df[spotify_df['Position'] <=10]

top10_songs.sort_values(by = ['Position', 'Track Name'], ascending = [False, True])[['Track Name','Position']].drop_duplicates()
```
